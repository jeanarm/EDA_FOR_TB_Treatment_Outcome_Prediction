### Given that the missing value is now removed, i can perform feature selection using statistical method, to only perform EDA on feature that can explain my target variable

```python
import numpy as np
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Assuming test_df is your DataFrame containing the data
X = test_df.drop(columns="Treatment outcome", axis=1)
y = test_df['Treatment outcome']

le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Identify numerical and categorical columns
numerical_cols = X.select_dtypes(include=np.number).columns
categorical_cols = X.select_dtypes(include=['object', 'category']).columns

# Apply one-hot encoding to categorical columns
encoder = OneHotEncoder(drop='first', sparse=False)
X_categorical_encoded = encoder.fit_transform(X[categorical_cols])
categorical_feature_names = encoder.get_feature_names_out(categorical_cols)

# Combine encoded categorical features with numerical features
X_encoded = pd.DataFrame(np.hstack((X[numerical_cols], X_categorical_encoded)), columns=np.concatenate((numerical_cols, categorical_feature_names)))

# Select top K features based on ANOVA F-value between features and target
k_features = 20  # Number of features to select
selector = SelectKBest(score_func=f_classif, k=k_features)
X_selected = selector.fit_transform(X_encoded, y_encoded)

# Get indices of selected features
selected_indices = selector.get_support(indices=True)
selected_feature_names = X_encoded.columns[selected_indices]

# Calculate ANOVA F-scores
f_scores = selector.scores_[selected_indices]

# true_f_scores = selector.scores_
# print("f_score",true_f_scores)
# Create a dictionary to associate feature names with F-scores
feature_scores = {feature: score for feature, score in zip(selected_feature_names, f_scores)}

# Sort features by F-scores in descending order
sorted_features = sorted(feature_scores.items(), key=lambda x: x[1], reverse=True)

# Print the sorted features and their scores
print("Sorted Features:")
for feature, score in sorted_features:
    print(f"{feature}: {score}")
```

## Model data transformation and model training 

```python
# Modelling
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression, RidgeClassifier, Lasso
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
X= loaded_df.drop(columns=['Treatment outcome'], axis=1)
y= loaded_df['Treatment outcome']
```

```python

# Select numerical and categorical features
num_features = X.select_dtypes(include=np.number).columns
cat_features = X.select_dtypes(include=np.object).columns

# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Preprocessing for features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_features),
        ('cat', OneHotEncoder(), cat_features)
    ])

X_preprocessed = preprocessor.fit_transform(X)

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y_encoded, test_size=0.2, random_state=42)

# Model Selection and Training
classifiers = {
    'KNeighbors': KNeighborsClassifier(),
    'DecisionTree': DecisionTreeClassifier(),
    'RandomForest': RandomForestClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'SVM': SVC(),
    'LogisticRegression': LogisticRegression(),
    'Ridge': RidgeClassifier()
    
}
print("Encoded Classes:", label_encoder.classes_)
```

```python
for name, classifier in classifiers.items():
    classifier.fit(X_train, y_train)
    y_pred_encoded = classifier.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred_encoded)
    precision = precision_score(y_test, y_pred_encoded, average='weighted')
    recall = recall_score(y_test, y_pred_encoded, average='weighted')
    f1 = f1_score(y_test, y_pred_encoded, average='weighted')
    
    print(f'{name}:')
    print(f'Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1: {f1:.4f}')
```
