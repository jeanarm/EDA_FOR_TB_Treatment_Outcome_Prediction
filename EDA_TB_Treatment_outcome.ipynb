## Read data set for jan,march and may 2020

```python
data_jan= pd.read_csv("Data Armel/L5 Jan2020 enrollments.csv")
data_jan['Enrollment date(Registration Date)'] = pd.to_datetime(data_jan['Enrollment date(Registration Date)'])
data_jan['Incidence date'] = pd.to_datetime(data_jan['Incidence date'])
data_jan
```

```python
data_march = pd.read_csv("Data Armel/L5 March2020 enrollments.csv")
data_march['Enrollment date(Registration Date)'] = pd.to_datetime(data_march['Enrollment date(Registration Date)'])
data_march['Incidence date'] = pd.to_datetime(data_march['Incidence date'])
data_march
```

```python
data_may = pd.read_csv("Data Armel/L5 May2020 enrollments.csv")
data_may['Enrollment date(Registration Date)'] = pd.to_datetime(data_may['Enrollment date(Registration Date)'])
data_may['Incidence date'] = pd.to_datetime(data_may['Incidence date'])
data_may
```

```python
## conbine 3 data sets on row basis so that we can merge the final data set with ready to merge data set above

concatenated_3months_df = pd.concat([data_jan, data_march,data_may], ignore_index=True)
concatenated_3months_df.sort_values(by='Enrollment date(Registration Date)')
concatenated_3months_df
```

## Merge 2 data set to make final data set to use for our project


```python
merged_df = pd.merge(ready_to_merge_df, concatenated_3months_df, on='Enrollment')
merged_df
```

```python
merged_df = pd.merge(ready_to_merge_df, concatenated_3months_df, on='Enrollment', how='outer')
merged_df
```

```python
## as you can see above, the two data sets had same columns and are duplicated which we dont want, lets merge them using combine_first method

#Speciying index that will help us to merge for both 2 data frames
df1 = ready_to_merge_df.set_index('Enrollment')
df2 = concatenated_3months_df.set_index('Enrollment')

#Merge them with combine_first method
final_df = df1.combine_first(df2)
final_df
```
